```python
# first coding exercise

def pad_zeros(arr, new_length):
    arr = np.pad(arr, (0, max(0, new_length - len(arr))), 'constant', constant_values=(0,0))
    return arr

print(pad_zeros(np.array([1,2,3,4]), 6))

# second coding exercise

def sum_arrays(arr_list):
  max_len = 0
  for i in range(len(arr_list)):
    if max_len < len(arr_list[i]):
      max_len = len(arr_list[i])

  result = np.zeros(max_len)
  for i in range(len(arr_list)):
    arr_list[i] = pad_zeros(arr_list[i], max_len)
    result += arr_list[i]

  return result

arr_list = [
np.array([2, 2, 4, 5, 4]),
np.array([1, 3, 2, 2]),
np.array([0, 4, 4, 2, 3]),
np.array([4, 5, 5])
]

print(sum_arrays(arr_list)) 
```

Exercise 3.1.2

```python
# first coding exercise

def getMedian(num_list):
    num_list.sort()
    return num_list[len(num_list)//2]
 
getMedian([3.12, 3.52, 3.4, 2.78, 3.68])

# second coding exercise
from math import sqrt
def calc_standard_deviation(num_list):
  avg = sum(num_list)/len(num_list)
  standard_deviation_list = [(num_list[i] - avg)**2 for i in range(len(num_list))]
  return sqrt((1/len(num_list))*sum(standard_deviation_list))

print(calc_standard_deviation([13,8,5,2]))
print(calc_standard_deviation([84,81,80,79,76]))
```

```python
import numpy as np

survey_data = pd.read_csv("https://s3.eu-west-1.amazonaws.com/data.cyber.org.il/virtual_courses/introdata/colab/youth_survey_preprocessed.csv")

# first coding exercise

print("Mean: " + str(survey_data["Height"].mean()) + " Median: " + str(survey_data["Height"].median()) +
      " Standard Deviation: " + str(survey_data["Height"].std()) + " Range: " + str(max(survey_data["Height"]) - min(survey_data["Height"])), end="\n\n")

# second coding exercise

print(survey_data["Punctuality"].value_counts(normalize=True), end="\n\n")

# third coding exercise

print(survey_data.loc[survey_data["Reliability"] >= survey_data["Reliability"].mean()]["Punctuality"].value_counts(normalize=True))

# fourth coding exercise

df_male = survey_data.loc[survey_data["Gender"] == "male"]
df_female = survey_data.loc[survey_data["Gender"] == "female"]
print(np.histogram(df_male["Height"], bins=range(145,206,10)))
print(np.histogram(df_female["Height"], bins=range(145,206,10)))
```
