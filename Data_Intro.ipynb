```python
import pandas as pd

# first coding exercise

series_of_strings = pd.Series(['this', 'is', 'a', 'Pandas', 'Series'])
series_of_strings[2:] = 'bla'
print(series_of_strings, end="\n\n")

# second coding exercise

series_of_strings = pd.Series(['this', 'is', 'a', 'Pandas', 'Series'])
print(series_of_strings.iloc[[3,1,2,3,4]], end="\n\n")

# third coding exercise

series_of_strings = pd.Series(['this', 'is', 'a', 'Pandas', 'Series'])
series_of_strings.index = ["First", "Second", "Second", "Second", "Fifth"]
print(series_of_strings.loc["Second"])
```

```python
import pandas as pd

# first coding exercise

top_five_popular = artists.sort_values(['popularity', 'followers'], ascending=False).iloc[0:5]["name"]
print(top_five_popular, end="\n\n")

# second coding exercise

top_five_popular = artists.sort_values(['popularity', 'followers'], ascending=False).iloc[0:5].sort_values('genres')["name"]
print(top_five_popular, end="\n\n")

# third coding exercise

print(artists.head(18).tail(2), end="\n\n")

# fourth coding exercise

hebrew_names = artists[['name', 'followers']].loc[(artists.name >= 'א') & (artists.name <= 'תתתת'), :].sort_values('followers').iloc[-1::-1]
print(hebrew_names, end="\n\n")

# fifth coding exercise 
exercise_five_list = artists.loc[(artists['popularity'] == 46)]
exercise_five_list = exercise_five_list.loc[exercise_five_list['followers'] >= 105000].loc[exercise_five_list['followers'] < 107000]
print(exercise_five_list.head()['name'])
```
